# Pandas, Histograms

```python
cols = ['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']

fig = plt.figure(figsize=(6,20))
for i in range(1,9):
    ax = fig.add_subplot(8,1,i)
    ax = recent_grads[cols[i-1]].hist(bins=20, xrot=40)
    ax.set_title(cols[i-1])
    
    
```

```python
recent_grads['Median'].hist(bins=20)
```

年俸の中央値は$30,000~35,000が最頻である

```python
recent_grads['ShareWomen'].hist(bins=50)
```

女性が96~98%を占める専攻分野がある一方で、男性が98%以上を占める専攻分野もあることがわかる。
